En Python, et donc aussi pour sympy, a mod n est donné par a % n.

Affichez les tables de multiplication de ${\mathbb Z}/12{\mathbb Z}$ et de ${\mathbb Z}/13{\mathbb Z}$.
Quels sont les éléments inversibles (et leurs inverses) dans les deux cas ?

```python
from sympy import *

def tabmul(n):
    print(' * |', end=' ')
    for y in range(1,n): 
        print ("%2d "% (y % n),end=' ') 
    print()
    print('-'*(4*n-1))
    for x in range(1,n): 
        print('%2d |'%x, end= ' ')
        for y in range(1,n): 
            print ("%2d "% ((x*y) % n),end=' ') 
        print()
```

a) Écrivez une fonction xgcd(a,b) qui renvoie un triplet $(d,u,v)$, où $d$ est le pgcd de $a,b$ et 
$d=ua+vb$ (coefficients de Bézout).

```python
def xgcd(a, b):
    if b == 0:
        return (a, 1, 0)
    else:
        (d, u, v) = xgcd(b, a % b)
        return (d, v, u - (a//b)*v)
```

b) Écrivez une fonction inversemod(a,n) qui calcule l'inverse de $a$ modulo $n$ s'il existe (et renvoie None sinon).
</ol>

```python
def inversemod(a, n):
    for i in range(1, n):
        if((a * i) % n == 1):
            return i
```

### 3. Petit théorème de Fermat

Affichez le triangle de Pascal jusqu'à $n=15$, successivement modulo 2,3,4,5. Où voyez-vous des zéros ?

```python
def binomial(n,k):
    if n<k: return 0
    if k==0: return 1
    return binomial(n-1,k-1)+binomial(n-1,k)


def triangle(n,r):
    for m in range(n+1):
        print ("%2d" % m, end = ' ')
        for k in range(m+1): 
            print ("%3d" % (binomial(m,k) % r),end = ' ')
        print()
```

```python
triangle(15,2)
```

```python
triangle(15,3)
```

```python
triangle(15,4)
```

```python
triangle(15,5)
```
